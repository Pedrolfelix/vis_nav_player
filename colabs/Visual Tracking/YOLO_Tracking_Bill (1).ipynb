```python
from ultralytics import YOLO

model = YOLO('yolov10s.pt')  # downloads pretrained weights automatically
```

```python
# Upload your video to Colab first, then set the path here
video_path = '/content/Bill.mp4'

results = model.track(
    source=video_path,
    classes=[0],           # 0 = person in COCO — no custom training needed
    conf=0.3,
    save=True,
    tracker='botsort.yaml'
)
```
